##1-2 **Check Permutation**: Given two strings, write a method to decide if one is a permutation of the other.

sol1: sort and compare

sol2: using table, compare the frequency of each char




```
#sol1: hash table, time: O(n), space: O(n)
def getFreq(test):
  table = {}
  for char in test:
    if char in table:
      table[char] += 1
    else:
      table[char] = 1
  return table

def sol1(str1, str2):
  if len(str1) != len(str2):
    return False
  table1 = getFreq(str1)
  table2 = getFreq(str2)

  for key, value in table1.items():
    if key not in table2 or table2[key] != value:
      return False
  return True

#sol1_enhance, time: O(n), space: O(n)
def sol1_enhance(str1, str2):
  if len(str1) != len(str2):
    return False
  table = getFreq(str1)

  for char in str2:
    if char not in table:
      return False
    table[char] -= 1
    if table[char] < 0:
      return False
  return True

#sol2: sort and compare, time:(nlogn), space: O(1)
def sol2(str1, str2):
  if len(str1) != len(str2):
    return False
  list1 = sorted(str1)
  list2 = sorted(str2)

  for idx in range(len(list1)):
    if list1[idx] != list2[idx]:
      return False
  return True

if sol2("aabbcc", "abcabc"):
  print(True)
else:
  print(False)
```

(Note: If implementing in Java, please use a character array so that you can perform this operation in place.)

sol2: in place copy，insert from final_index, but using real_index to traverse


```
# sol1, copy to other string, time:O(n), space: O(n)
# sol2, in place insert, time:O(n), space: O(1)
def sol2(string, real_length):
  final_index = len(string)-1
  slist = list(string)
  for idx in range(real_length-1, -1, -1):
    if string[idx] == " ":
      slist[final_index] = '0'
      final_index -=1
      slist[final_index] = '2'
      final_index -=1
      slist[final_index] = '%'
      final_index -=1
    else:
      slist[final_index] = string[idx]
      final_index -= 1
  return ''.join(slist)

string = sol2('Mr John Smith    ', 13)
print(string)
```

##1-5 **One Away**: There are three types of edits that can be performed on strings: insert a character, remove a character, or replace a character. Given two strings, write a function to check if they are one edit (or zero edits) away.

sol: seperate replace, insert, remove condition

```
# sol: seperate replace, insert, remove condition, time:O(n), space: O(1)
def checkReplace(string1, string2):
  if len(string1) != len(string2):
    return False
  count = 0
  for idx in range(len(string1)):
    if string1[idx] != string2[idx]:
      count += 1
    if count > 1:
      return False
  return True

def checkDiff(short_string, long_string):
  idx_short = idx_long = 0
  diff = 0
  while idx_short < len(short_string):
    if short_string[idx_short] != long_string[idx_long]:
      idx_long += 1
      diff += 1
    if diff > 1:
      return False
    idx_long += 1
    idx_short += 1
  return True


def sol(string1, string2):
  if len(string1) == len(string2):
    return checkReplace(string1, string2)
  elif len(string1) - len(string2) == 1:
    return checkDiff(string2, string1)
  elif len(string2) - len(string1) == 1:
    return checkDiff(string1, string2)
  else:
    return False

if sol('pale', 'zales'):
  print(True)
else:
  print(False)
```

##1-6 **String Compression**: Implement a method to perform basic string compression using the counts of repeated characters. For example, the string aabcccccaaa would become a2b1c5a3. If the "compressed" string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a - z).

sol: iteration

```
#sol: iteration, time: O(n), space: O(n)
def sol(string):
  result = ""
  if len(string) == 0:
    return string
  count = 0
  for idx in range(len(string)):
    cur = string[idx]
    count += 1
    if idx+1 == len(string) or string[idx+1] != cur:
      result += cur + str(count)
      count = 0
  return string if len(result) > len(string) else result

print(sol('aabcccccaaa'))
```

##1-7 **Rotate Matrix**: Given an image represented by an NxN matrix, where each pixel in the image is 4 bytes, write a method to rotate the image by 90 degrees. Can you do this in place?
*Leetcode 48*

sol: just rotate, set boudries in each iteration, time: O($n^2$), space: O(1)

```
def sol(matrix):
  if not matrix:
    return
  rowup = 0
  rowdown = len(matrix)-1
  colleft = 0
  colright = len(matrix[0])-1

  while rowup < rowdown:
    for col in range(colleft, colright):
      offset = col-colleft
      tmp = matrix[rowup][col]
      matrix[rowup][col] = matrix[rowdown-offset][colleft]
      matrix[rowdown-offset][colleft] = matrix[rowdown][colright-offset]
      matrix[rowdown][colright-offset] = matrix[rowup+offset][colright]
      matrix[rowup+offset][colright] = tmp

    rowup += 1
    rowdown -= 1
    colleft += 1
    colright-= 1

matrix = [[1,2], [3, 4]]
print(matrix)
sol(matrix)
print(matrix)
```

```
#sol1: using extra list recording zero row and zero col, time:O(n^2), space: O(n)
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        if not matrix:
            return matrix
        zero_row_list = []
        zero_col_list = []
        
        for row in range(len(matrix)):
            for col in range(len(matrix[0])):
                if matrix[row][col] == 0:
                    zero_row_list.append(row)
                    zero_col_list.append(col)
        for row in zero_row_list:
            matrix[row] = [0]*len(matrix[0])
        for row in range(len(matrix)):
            for col in zero_col_list:
                matrix[row][col] = 0

#sol2: using existing space, time: O(n^2), space: O(1)
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        if not matrix:
            return matrix
        first_row_has_zero = False
        first_col_has_zero = False
        
        for row in range(0, len(matrix)):
            for col in range(0, len(matrix[0])):
                if row == 0 and matrix[row][col] == 0:
                    first_row_has_zero = True
                if col == 0 and matrix[row][col] == 0:
                    first_col_has_zero = True

                if matrix[row][col] == 0:
                    matrix[0][col] = 0
                    matrix[row][0] = 0
        
        for row in range(1, len(matrix)):
            if matrix[row][0] == 0:
                matrix[row] = [0]*len(matrix[0])
                
        for col in range(1, len(matrix[0])):
            if matrix[0][col] == 0:
                for row in range(1, len(matrix)):
                    matrix[row][col] = 0
        
        if first_row_has_zero:
            matrix[0] = [0]*len(matrix[0])
        if first_col_has_zero:
            for row in range(0, len(matrix)):
                matrix[row][0] = 0
```

```
#sol: two pointers, time:(max(length)), space:O(1)
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        if not headA or not headB:
            return None
        lengthA = lengthB = 0
        pA = headA
        pB = headB
        
        while pA != None or pB != None:
            if pA:
                lengthA += 1
                pA = pA.next
            if pB:
                lengthB += 1
                pB = pB.next
        if pA != pB:
            return None
        
        pA = headA
        pB = headB
        if lengthA > lengthB:
            for idx in range(lengthA-lengthB):
                pA = pA.next
        elif lengthB > lengthA:
            for idx in range(lengthB-lengthA):
                pB = pB.next
        while pA != pB:
            pA = pA.next
            pB = pB.next
        return pA
        
    
    def getLength(self, node):
        if not node:
            return 0
        return 1 + self.getLength(node.next)
        
```

```
class Stack:
    def __init__(self, size):
        self.top = None
        self.capacity = size
        self.stack = [] * size
    def __init__(self):
        self.top = None
        self.capacity = sys.maxsize
        self.stack = []
    def push(self, value):
        if (self.size() == self.capacity):
            return False

        self.stack.append(value)
        if (self.size() == 0):
            self.top = 0
        else:
            self.top = self.top + 1
        return True
    def pop(self):
        if (self.size() == 1):
            self.top = None
        else:
            self.top = self.top - 1
        return self.stack.pop()

    def size(self):
        if (self.top == None):
            return 0
        return self.top + 1
    def isEmpty(self):
        return self.size() == 0

    def peek(self):
        if (self.size() == 0):
            return None
        return self.stack[self.top]
    def print_data(self):
        print(self.stack)
```

```
#sol3
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        if self.isCover(p, q):
            return p
        if self.isCover(q, p):
            return q
        return self.helper(root, p, q)
        
        
    def helper(self, node, p, q):
        if not node:
            return None
        p_in_left = self.isCover(node.left, p)
        q_in_left = self.isCover(node.left, q)
        if p_in_left and q_in_left:
            return self.helper(node.left, p, q)
        elif not p_in_left and not q_in_left:
            return self.helper(node.right, p, q)
        else:
            return node
        
    def isCover(self, node, target):
        if not node:
            return False
        if node == target:
            return True
        return self.isCover(node.left, target) or self.isCover(node.right, target)

#sol4 return node and is_ancestor
class Ret:
    def __init__(self, node, is_ancestor):
        self.node = node
        self.is_ancestor = is_ancestor
class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        if not root:
            return None
        ret = self.helper(root, p, q)
        return ret.node if ret.is_ancestor else None
    
    def helper(self, node, p, q):
        if not node:
            return Ret(None, False)
        if node == p and node == q:
            return Ret(node, True)
        
        left_ret = self.helper(node.left, p, q)
        right_ret = self.helper(node.right, p, q)
        
        if left_ret.is_ancestor:
            return left_ret
        if right_ret.is_ancestor:
            return right_ret
        if left_ret.node and right_ret.node:
            return Ret(node, True)
        
        if node == p or node == q:
            is_ancestor = left_ret.node or right_ret.node
            return Ret(node, is_ancestor)
        if left_ret.node:
            return left_ret
        if right_ret.node:
            return right_ret
        return Ret(None, False)
```

```
class Solution:
    def floodFill(self, image: List[List[int]], sr: int, sc: int, newColor: int) -> List[List[int]]:
        self.helper(image, sr, sc, image[sr][sc], newColor)
        return image
        
    def helper(self, image, row, col, old_color, new_color):
        if not image:
            return
        if row < 0 or row >= len(image):
            return
        if col < 0 or col >= len(image[0]):
            return
        if old_color == new_color:
            return
        if image[row][col] != old_color:
            return
        image[row][col] = new_color
        self.helper(image, row-1, col, old_color, new_color)
        self.helper(image, row+1, col, old_color, new_color)
        self.helper(image, row, col-1, old_color, new_color)
        self.helper(image, row, col+1, old_color, new_color)
```

## 8.12 **Eight Queens**: Write an algorithm to print all ways of arranging eight queens on an 8x8 chess board so that none of them share the same row, column, or diagonal. In this case, "diagonal" means all diagonals, not just the two that bisect the board.

*Leetcode51*

sol1: 
* 從row為0出發，for迴圈放每個col為出發，如果可以放成功，就append座標到collect裡，繼續下一列
* Backtracking，當row到為girdsize的時候，就是可以蒐集結果的時候，所以再塞進results裡，
* 而每一層的colum擺放位置都用cols紀錄。
* 當row_diff == col_diff時，就是所謂的對角線，不能擺放。



```
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        if n == 0:
            return []
        collect = []
        result = []
        self.helper(0, n, collect, result)
        return result
        
    def helper(self, row, size, collect, result):
        if row == size:
            result.append(self.genMap(collect))
            return
        for col in range(size):
            if not self.isValid(collect, col):
                continue
            collect.append(col)
            self.helper(row+1, size, collect, result)
            collect.pop()
            
    def isValid(self, collect, current_col):
        for row, col in enumerate(collect):
            #collumn validate
            if col == current_col:
                return False
            #diag validate
            current_row = len(collect)
            row_diff = abs(current_row - row)
            col_diff = abs(current_col - col)
            if row_diff == col_diff:
                return False
        return True
    
    def genMap(self, collect):
        ret = []
        for row, col in enumerate(collect):
            tmp = ['.' for i in range(len(collect))]
            tmp[col] = 'Q'
            ret.append("".join(tmp))
        return ret
```

```
#Bubble sort
def Bubble_sort(array):
  for end in range(len(array)-1, -1, -1):
    for idx in range(end-1):
      if array[idx] > array[idx+1]:
        swap(array, idx, idx+1)
  return array

def swap(array, a, b):
  tmp = array[a]
  array[a] = array[b]
  array[b] = tmp

print(Bubble_sort([2,3,4,1,5,7,6,10]))
```

```
#Selection sort
import sys

def selection_sort(array):
  for idx in range(len(array)):
    minimum = sys.maxsize
    minimum_idx = -1
    for idy in range(idx, len(array)):
      if array[idy] < minimum:
        minimum = array[idy]
        minimum_idx = idy
    swap(array, idx, minimum_idx)

array = [2,3,4,1,5,7,6,10]
selection_sort(array)
print(array)
```

```
#Insertion sort

def insertion_sort(array):
  for idx in range(1, len(array)):
    for idy in range(idx, 0, -1):
      if array[idy] < array[idy-1]:
        swap(array, idy, idy-1)
      else:
        break
  return array


array = [2,3,4,1,5,7,6,10]
insertion_sort(array)
print(array)
```

```
#Merge Sort

def merge_sort(array, start, end):
  if start >= end:
    return [array[start]]
  mid = start + (end-start)//2
  a = merge_sort(array, start, mid)
  b = merge_sort(array, mid+1, end)
  return merge(a, b)

def merge(array_a, array_b):
  idx_a = 0
  idx_b = 0
  result = []
  while idx_a < len(array_a) or idx_b < len(array_b):
    if idx_a == len(array_a):
      result.append(array_b[idx_b])
      idx_b += 1
    elif idx_b == len(array_b):
      result.append(array_a[idx_a])
      idx_a += 1
    elif array_a[idx_a] <= array_b[idx_b]:
      result.append(array_a[idx_a])
      idx_a += 1
    else:
      result.append(array_b[idx_b])
      idx_b += 1
  return result

array = [2,3,4,1,5,7,6,10]
result = merge_sort(array, 0, len(array)-1)
print(result)
```

```
#Quick sort
import random

def quick_sort(array, start, end):
  if start >= end:
    return [array[start]]
  pivot_idx = random.randint(start, end)
  pivot = array[pivot_idx]
  swap(array, pivot_idx, end)
  i = start-1
  j = start
  while j < end:
    if array[j] < pivot:
      i += 1
      swap(array, i, j)
    j += 1
  i += 1
  swap(array, i, end)
  quick_sort(array, i+1, end)
  quick_sort(array, start, i-1)


array = [2,3,4,1,5,7,6,10]
quick_sort(array, 0, len(array)-1)
print(array)
```

* Sol: 先用index為1，然後每次*2去找到為2147483647 或是 比search_target大的時候停止。接下來就是binary search了
  * start == index//2, end == index，用start, end 算出mid，當array[mid] == 2147483647的時候，要向左找，一樣讓high = mid-1即可。


```
#time: O(logn)
class Solution:
    def search(self, reader, target):
        """
        :type reader: ArrayReader
        :type target: int
        :rtype: int
        """
        idx = 1
        while reader.get(idx) != 2147483647 and reader.get(idx) < target:
            idx *= 2

        start = idx//2
        end = idx
        while start <= end:
            mid = start + (end-start)//2
            if reader.get(mid) == target:
                return mid
            elif reader.get(mid) > target or reader.get(mid) == 2147483647:
                end = mid-1
            else:
                start = mid+1
        return -1
```

對角線去找，但是要注意矩形的情況，也就是每次的recursive，除了固定參數傳進來的start和end，要多兩個變數專門traverse對角線用的，要先算長寬一制的diff，才可以處理矩形問題，因為只有正方形才有對角線。接著就是對角線的bineary search，一直while到leftup > rightdown
接著做partition，要把start跟end代進去，並且leftup也要一起，leftup剛好會落在比要找的數在大一點的位置，所以可以用leftup, start, end的位置去算去左下跟右上的partition各對角線的start 跟 end，然後繼續


```
#sol3
class Coordinate:
    def __init__(self, row, col):
        self.row = row
        self.col = col
    def InBound(self, matrix):
        return self.row < len(matrix) and self.row >=0 and self.col < len(matrix[0]) and self.col >=0
    def IsBefore(self, coordinate):
        return self.row <= coordinate.row and self.col <= coordinate.col

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        if not matrix:
            return False
        start = Coordinate(0, 0)
        end = Coordinate(len(matrix)-1, len(matrix[0])-1)
        return self.helper(matrix, start, end, target)
    def helper(self, matrix, start, end, target):        
        if not start.InBound(matrix) or not end.InBound(matrix):
            return False
        if not start.IsBefore(end):
            return False
        diff = min(end.row-start.row, end.col-start.col)
        leftup = Coordinate(start.row, start.col)
        rightdown = Coordinate(start.row+diff, start.col+diff)
        while leftup.IsBefore(rightdown):
            mid = Coordinate(leftup.row+(rightdown.row-leftup.row)//2, leftup.col+(rightdown.col-leftup.col)//2)
            if matrix[mid.row][mid.col] == target:
                return True
            elif matrix[mid.row][mid.col] > target:
                rightdown = Coordinate(mid.row-1, mid.col-1)
            else:
                leftup = Coordinate(mid.row+1, mid.col+1)
        return self.partition(matrix, start, end, leftup, target)
    def partition(self, matrix, start, end, leftup, target):
        right_up_start = Coordinate(start.row, leftup.col)
        right_up_end = Coordinate(leftup.row-1, end.col)
        left_down_start = Coordinate(leftup.row, start.col)
        left_down_end = Coordinate(end.row, leftup.col-1)
        return self.helper(matrix, right_up_start, right_up_end, target) or self.helper(matrix, left_down_start, left_down_end, target)
        
```

##10-11 **Peaks and Valleys**: In an array of integers, a "peak" is an element which is greater than or equal to the adjacent integers and a "valley" is an element which is less than or equal to the adjacent integers. For example, in the array {5, 8, 6, 2, 3, 4, 6}, {8, 6} are peaks and {5, 2} are valleys. Given an array of integers, sort the array into an alternating sequence of peaks and valleys.

*EXAMPLE*
Input: {5, 3, 1, 2, 3} Output: {5, 1, 3, 2, 3}

* sol1:先排序整個array，因為是small <= middle <= large，所以拿middle跟small互換，一定會有valley跟peak出現，一次跳兩個，從idx為1開始，可以讓code比較精簡. time:O(nlogn)
* sol2: 最大放中間，一樣是從idx為1開始，並且一次跳兩個。比較左右，找出最大值所代表的idx，如果有找到跟mid互換。 If we're swapping middle with left, then left is currently a valley. Middle is smaller than left, so we're putting an even smaller element as a valley. Nothing will break. All is good!
如果right == len(array)時，就用-sys.maxint-1但替他們的值，這樣code比較精簡。time:O(n)


```
import sys
#time: O(nlogn)
def sol1(array):
  array = sorted(array)
  for idx in range(1, len(array), 2):
    tmp = array[idx-1]
    array[idx-1] = array[idx]
    array[idx] = tmp
  return array

#time: O(n)
def swap(array, idx_a, idx_b):
  tmp = array[idx_a]
  array[idx_a] = array[idx_b]
  array[idx_b] = tmp

def sol2(array):
  for idx in range(1, len(array), 2):
    left = array[idx-1]
    mid = array[idx]
    right = array[idx+1] if idx+1 < len(array) else -sys.maxsize-1
    
    max_value = max(left, mid, right)
    if max_value == left:
      swap(array, idx-1, idx)
    elif max_value == right:
      swap(array, idx+1, idx)
  return array


print(sol1([5,3,1,2,3]))
print(sol2([5,3,1,2,3]))
```
